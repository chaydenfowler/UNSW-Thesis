## 2.2.1  Multi Import Attempt
Following from unswDataImportSingle.ipynb, we now are going to try import the entirety of the 2018-Array folder. We can select our chosen headers using a list comprehension, choosing to keep "Timestamp, TmpAmb, WindVel, and A.Ms.Watt" as (0, 4, 6, 9).
Using glob.iglob, we can create an iterable for the path of all the relevant files in the 2018-Array folder. By specifying in pd.read_csv the usecols=headers, we make sure to only import the relevant headers (and ignoring the other 357 that we don't want) which reduces import time and memory usage. After this we can set the indices to the relevant datetime that the data was captured at, making sure we get a datetime64 result for later manipulation. I've again truncated the headers string to only the first 10 or so elements to reduce its space. 

```python
import pandas as pd
import glob
import os

headers = "TimeStamp;ExlSolIrr;IntSolIrr;SMA-h-On;TmpAmb C;TmpMdul C;WindVel km/h;A.Ms.Amp;A.Ms.Vol;A.Ms.Watt;A1.Ms.Amp"
headers = dict(enumerate(headers.split(';')))
headers = {k: headers[k] for k in (0, 4, 6, 9)} # choosing which headers we want
    
path = r"C:\Users\Clairvoyant Cabbage\Documents\PythonProject\Thesis\UNSWData\2018-Array"
all_files = glob.iglob(os.path.join(path, "*csv"))
df1 = pd.concat((pd.read_csv(f, delimiter=";", header=None, skiprows=6, usecols=headers).assign(filename = os.path.basename(f)) for f in all_files))
df1 = df1.rename(columns = headers)
df1['TimeStamp'] = df1['filename'].str.split('.', expand = True)[0] + " " + df1['TimeStamp'] #setting timestamp to proper timestamp
df1['TimeStamp'] = pd.to_datetime(df1['TimeStamp']) #setting timestamp to a datetime64[ns] type
df1.index = df1['TimeStamp'] #setting the index to the timestamp
df1 = df1.drop(columns = ['filename']) #dropping filename as we don't need it anymore
df1.info()
```

## 2.2.2 Result
As a result we can see we get the datetime64\[ns\] type required for our timestamp (this is so pandas will be nice to us), and we have no irregular data types. The memory usage appears to have been correctly managed by us giving us only 3.3MB of data stored, whilst each individual file imported is ~0.3 MB each (for ~365 files per year) meaning we're only storing about 3% of the imported data. We know that there is definitively missing data as shown by the non-null counts in each column; A.Ms.Watt aka Power has the lowest count as expected as we should not be recording during night hours and this shouldn't pose any issue. The section of power records sitting at zero for the entirety of a month may be problematic, but should be solvable. The missing data on Ambient temperature and Wind velocity seem to predominantly be on one missing chunk as seen in the plots below. It is concerning to have a large section of data missing, however missing data points can otherwise be imputed.

```python
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
%matplotlib inline

# plot the AC power time series, Temp, Timestamp and Wind velocity
fig, ax = plt.subplots(3, figsize = (10,10))
plt.subplots_adjust(hspace = 0.2)
ax[0].plot(df1.index, df1['A.Ms.Watt'], 'o', alpha = 0.01)
ax[0].set_ylim(0,7000)
ax[0].set_ylabel('AC Power (W)');
ax[1].plot(df1.index, df1['TmpAmb C'], 'o', alpha = 0.01)
ax[1].set_ylim(0,50)
ax[1].set_ylabel('Ambient Temperature (C)');
ax[2].plot(df1.index, df1['WindVel km/h'], 'o', alpha = 0.01)
ax[2].set_ylim(0,50)
ax[2].set_ylabel('Wind Velocity (km/h)');
```
