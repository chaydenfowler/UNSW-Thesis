## 2.1.1 Import Attempt
Here is the first data import test on singular TEBT file, using the 01/01/2018 dated file.
For this file we are forced to skip the first 6 rows (which is when the actual data starts), and to ignore the headers.
The original headers in the file are partially uncomaptable due to an encoding issue (pandas doesn't like the degree symbol in the encoding it chose) and despite a true solution being avaialble, the workaround is to specify all the headings hardcoded. 
Once we have all the data we can rename the headers to the original headers by enumerating the header list. We've truncated the headings string to include only the 10 or so headings relavated before shortening it further; the full 362 headings can be found in the source data csv.

```python
import pandas as pd
file_name = r"C:\Users\Clairvoyant Cabbage\Documents\PythonProject\Thesis\UNSWData\2018-Array\2018-01-01.csv"
headers = "TimeStamp;ExlSolIrr;IntSolIrr;SMA-h-On;TmpAmb C;TmpMdul C;WindVel km/h;A.Ms.Amp;A.Ms.Vol;A.Ms.Watt;A1.Ms.Amp"
headers = dict(enumerate(headers.split(';')))
headers = {k: headers[k] for k in (0, 4, 6, 9)} # choosing which headers we want
df1 = pd.read_csv(file_name, delimiter=";", header=None, usecols=headers, skiprows=6)
df1 = df1.rename(columns = headers)
print(df1)
```

## 2.1.2 Result
We can see from the initial result that we have roughly 24 hours of data available, with 362 columns of data points to choose from of which we selected 4 relevant to the data we need. A significant portion of our data will be 'Not a Number' (NaN) if they're missing from the original CSV, but this wont pose many issues and is very common. Deciding which columns to keep may prove challenging and can be addressed later. Using a small bit of code to check what our power over the day looks like, along with our other data. We can see below that we mostly have a typical day, but for example maybe cloud cover over midday onwards until the evening. The Timestamp is linear and not showing any irregularities either, which means we can rely on our data to be imported in time order.

```python
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
%matplotlib inline

# plot the AC power time series, Temp, Timestamp and Wind velocity
fig, ax = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.4)
ax[0, 0].plot(df1.index, df1['A.Ms.Watt'], 'o', alpha = 0.1)
ax[0, 0].set_ylim(0,7000)
ax[0, 0].set_ylabel('AC Power (W)');
ax[0, 1].plot(df1.index, df1['TmpAmb C'], 'o', alpha = 0.1)
ax[0, 1].set_ylim(0,40)
ax[0, 1].set_ylabel('Ambient Temperature (C)');
ax[1, 0].plot(df1.index, df1['TimeStamp'], 'o', alpha = 0.1)
ax[1, 0].set_ylabel('TimeStamp');
ax[1, 0].tick_params('both', labelleft=False)
ax[1, 1].plot(df1.index, df1['WindVel km/h'], 'o', alpha = 0.1)
ax[1, 1].set_ylim(0,20)
ax[1, 1].set_ylabel('Wind Velocity (km/h)');
```
