```python
import pandas as pd
import glob
import os
import numpy as np
```

```python
path = r"C:\Users\chayd\Documents\Projects\UNSW-Thesis\Data"
all_files = glob.iglob(os.path.join(path, "*-PV/*.csv"))
#all_files = glob.iglob(os.path.join(path, "*12-PV/*-01-01.csv"))
#for item in all_files:
#    print(item)
    
headers = ["TimeStamp","TmpAmb C","WindVel km/h","305049607"]#,"2110144662","2110180596","2110180969",
           #"2110192945","2110018240","2110180932","2110183843","2110183847","2110183895","2110192909"]
cols =    [0,4,6,9]#,54,85,116,147,178,209,240,271,302,333]
headers = {k: header for k, header in zip(cols, headers)}

firstMerge = True
failedImports = 0
for f in all_files:
    try:
        dfTemp = pd.read_csv(f, delimiter=";", header=None, skiprows=6, usecols=cols).assign(filename = os.path.basename(f))
        dfTemp = dfTemp.rename(columns = headers)
        dfTemp['TimeStamp'] = dfTemp['filename'].str.split('.', expand = True)[0] + " " + dfTemp["TimeStamp"].str.split(" ").str[-1] #setting timestamp to proper timestamp
        dfTemp.index = pd.to_datetime(dfTemp['TimeStamp']) #setting index to a datetime64[ns] type
        dfTemp = dfTemp.drop(columns = ['filename', 'TimeStamp']) #dropping filename and TimeStamp as we don't need them anymore
        if firstMerge:
            dfPV = dfTemp
            firstMerge = False
        else:
            dfPV = pd.concat([dfPV, dfTemp])
            print(f.split("\\")[-1], " successful concat")
    except Exception as e:
        print(f.split("\\")[-1], e)
        failedImports += 1   
        
print("failed to import: {0} PV files\n".format(failedImports))
dfPV.info()
```
